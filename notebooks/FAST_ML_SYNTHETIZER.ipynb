```python
data=pd.read_csv("../data/cleaned_dataset.csv")
```

```python
metadata=data.info()
```

```python
from sdv.lite import TabularPreset
model = TabularPreset(name='FAST_ML')
model.fit(data)
```

```python
synthetic_data = model.sample(num_rows=10000)
```

```python
synthetic_data=synthetic_data.drop("Unnamed: 0",axis=1)
```

```python
synthetic_data
```

```python
import statsmodels.api as sm
X = synthetic_data.drop("basesalary", axis=1)
Y = synthetic_data["basesalary"]

regression = sm.OLS(Y, X).fit()
predictions = regression.predict(data.drop(["basesalary","Unnamed: 0"],axis=1))
regression.summary()
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error
rmse = float(format(np.sqrt(mean_squared_error(data["basesalary"], predictions)), '.3f'))
print("\nRMSE: ", rmse)
```
